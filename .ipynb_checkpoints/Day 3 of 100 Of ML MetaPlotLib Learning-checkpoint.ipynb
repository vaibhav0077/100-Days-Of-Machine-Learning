# Legend 

```python
days=[1,2,3,4,5,6,7]   # x axis number
max_t=[50,51,52,48,47,49,46] # y axis num
min_t=[43,42,40,44,33,35,37] # y axis num
avg_t=[45,48,48,46,40,42,41] # y axis num
```

# Axes Labels and  chart titles

```python
plt.plot(days, max_t, label="max") # x axis ,  y -axis , lables
plt.plot(days, min_t, label="min")
plt.plot(days, avg_t, label="average")
```

# Legends

```python
plt.plot(days, max_t, label='max')
plt.plot(days, min_t, label='min')
plt.plot(days, avg_t, label='avg')

plt.legend(loc='best', shadow=True, fontsize='large' ) 

plt.grid()
```

# Bar Chart

```python
plt.xticks(days, min_t)
plt.bar(days, max_t)
plt.bar(days, min_t)
plt.bar(days, avg_t)
```

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
profit= [50, 65, 40, 85]
```

```python
import numpy as np
xpos = np.arange(len(company))
xpos
```

```python
plt.bar(xpos-0.2,revenue, width=0.3, label="revenue" )
plt.xticks(xpos, company)
plt.ylabel("Revenu(cmp)")
plt.title("Company Revenus")

plt.bar(xpos+0.2, profit, width=0.3, label="Profit")
plt.legend(loc='best')
```

```python
plt.barh(xpos-0.2,revenue, label="revenue" )
plt.yticks(xpos, company)
plt.xlabel("Revenu(cmp)")
plt.title("Company Revenus")

plt.barh(xpos+0.2, profit,  label="Profit")
plt.legend(loc='best')
```

# Histogram

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_1 = [113, 85, 100, 150, 140, 88, 93, 15, 135, 80, 77, 82, 129]
# plt.hist(blood_sugar, rwidth=0.8) # by default number of bins is set to 10
# plt.hist(blood_sugar, rwidth=0.8, bins = 6)
# plt.hist(blood_sugar, rwidth=0.8, bins = [80, 100, 130, 200, 250], color="g", histtype='step')
plt.hist([blood_sugar, blood_sugar_1 ], rwidth=0.8, bins = [80, 100, 130, 200, 250], color=['green', 'red'], label=['men', 'women'], orientation="horizontal")
plt.legend(shadow="True")
```

# PieChart

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
```

```python
plt.pie(exp_vals, labels=exp_labels, radius=1.5, autopct ="%0.2f%%", shadow=True, explode=[0, 0.2, 0.3, 0.1, 0.4 ], startangle=45)
plt.show()

plt.savefig("piechart.png", bbox_inches="tight")
```
