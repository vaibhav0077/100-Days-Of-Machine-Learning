```python
X = df['state_code'].value_counts()
Y = (np.arange(df['state_code'].value_counts().max()) + 5)
print(X)
print(Y)
```

```python
import matplotlib.pyplot as plt
X.plot(kind='bar', figsize=(25,5), fontsize=20, legend=True, xlabel='state code', ylabel='number of cities')
# plt.xlabel('state code')
plt.ylabel('number of cities', fontsize=20)
plt.title("total number of cities in each states code", fontsize=35)
plt.xlabel("state code")
```
