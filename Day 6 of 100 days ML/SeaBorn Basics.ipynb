# Seaborn 

# Distribution plots

```python
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
df.sample(3)
```

```python
df.shape
```

```python
# Coorelation
df.corr()
```

```python
df.dtypes
```

```python
sns.heatmap(df.corr())
```

# JoinPlot

```python
sns.jointplot(x='tip', y='total_bill', data=df, kind='hex')
```

```python
sns.jointplot(x='tip', y='total_bill', data=df, kind='reg') # regression -> reg
```

# PairPlot

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue='sex') 
```

```python
sns.pairplot(df, hue='smoker') 
```

# DistPlot

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'], kde=False, bins=10)
```

# Categorical Plot

```python
sns.countplot('sex', data=df)
```

```python
sns.countplot('day', data=df)
```

```python
sns.countplot(y='sex', data=df)
```

```python
sns.countplot( y='sex', data=df)
```

```python
sns.barplot(x='sex', y='total_bill', data=df)
```

```python
sns.boxplot('smoker', 'total_bill', data=df)
```

```python
sns.boxplot('day', 'total_bill', data=df, palette='rainbow')
```

```python
sns.boxplot('day', 'total_bill', data=df, palette='rainbow', hue='smoker')
```

# Violine Plot

```python
sns.violinplot(x='total_bill', y='day', data=df, palette='rainbow')
```
